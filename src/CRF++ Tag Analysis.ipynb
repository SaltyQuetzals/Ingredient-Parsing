```python
from matplotlib import pyplot as plt
from open_seq_data import read_crf_file, TAG_MAP
from collections import Counter
from pprint import pprint as print
```

```python
dataset = read_crf_file('../data/train.crftags')
tokens, tags = zip(*dataset)

REVERSE_MAP = {v: k for k, v in TAG_MAP.items()}

all_tags = []
for tagset in tags:
    all_tags += [REVERSE_MAP[t] for t in tagset]

tag_counter = Counter(all_tags)
x, y = zip(*tag_counter.most_common())


fig = plt.figure(figsize=(12, 8))

plt.bar(x, y)
plt.title('Tag Frequencies')
print(tag_counter.most_common())
```
